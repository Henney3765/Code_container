```python
# import library
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 노트북에 플롯의 정적 이미지를 삽입
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
# plt.show는 지금까지의 그림을 그리기 위한 최종 지시를 의미
```

```python
np.random.seed(42)
y = np.random.standard_normal(30)
```

```python
x = range(len(y))
```

```python
plt.plot(x, y)
plt.show()
```

## subplot()를 이용하여 그림 분리하기

```python
plt.figure()
plt.subplot(2, 1, 1) # row, columns, panel number
plt.plot(x, np.cos(x))

# 두번 째 패널에 aixs 설정
plt.subplot(2, 1, 2)
plt.plot(x, np.sin(x))
plt.show()
```

## 객체지향 인터페이스 사용하기

```python
# subplot() 메소드를 사용하여 객체지향 인터페이스 방법으로 위와 동일한 결과를 출력
# plt.plot()에서 ax[].plot()으로 변경
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.cos(x))
ax[1].plot(x, np.sin(x))
plt.show()
```

## xlim(), ylim(): 그래프 x, y축 범위 축소

```python
plt.xlim(0, 10)
# plt.ylim(1, 2)
plt.plot(y.cumsum())
```

```python
# 색 약자
# b: 파란색 / r: 빨간색 / m: 자홍색 / k: 검정색
# g: 녹색 / c: 청록색 / y: 노란색 / w: 흰색

# 표준 스타일 약자
# p: 오각형 표시 / o: 원 표시 / .: 점 표시
# - : 실선 / -.: 대시-점선 / :: 점선 / v: 아래쪽 삼각형 표시 / d: 다이아 표시
```

```python
plt.figure(figsize=(15, 8))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'ks') # 검정색 사각형 표시 (색 약자 / 점 스타일 조합)
plt.grid(True) # 그림에 grid
plt.axis('tight') # 축 간격을 조밀하게
plt.xlabel('Index') # x 라벨
plt.ylabel('value') # y 라벨
plt.title('Example Plot') # title 설정
plt.show()
```
